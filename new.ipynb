```python
import numpy as np
arr = np.array([
    [1,2,3],
    [4,5,6],
])
arr.flatten(order = 'C')
arr = np.ravel(arr)
arr

```

```python
arr = np.array([[[[1,2,3]]]])
for i in np.nditer(arr):
    print(i)

for i,j in np.ndenumerate(arr):
    print(i,j)   
```
