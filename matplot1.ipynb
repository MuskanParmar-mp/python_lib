```python
avg_tip = df.groupby('day')['tip'].mean()
avg_tip
```

```python
plt.Figure(figsize=(5,3))
avg_tip.plot(kind='bar', colormap="Set2")
plt.ylabel("avg_tip")
plt.title("Avg  tips by Days")
```

```python
plt.Figure()

for size, group in df.groupby('size'):
    count = len(group)
    label_text = f"Size {size} (n = {count})"

    plt.scatter(group['total_bill'], group['tip'], label= label_text)


#plt.scatter(df['total_bill'], df['tip'], c=df['size'], cmap='viridis')
plt.legend()
plt.xlabel("total bill")
plt.ylabel("freq")
plt.title("Total bill by tips")
plt.show()
```

```python
plt.Figure()
scatter = plt.scatter(df['total_bill'], df['tip'], c=df['size'], cmap='coolwarm')
cbar = plt.colorbar(scatter)
cbar.set_label("Party Size")
```

```python
smoker_count = df.groupby('smoker')['size'].count()
smoker_count
```

```python
plt.Figure(figsize=(3,3))
#smoker_count.plot(kind='pie', colormap="Set2")
plt.pie(smoker_count.values,labels=smoker_count.index, autopct='%1.1f%%',shadow=True)

plt.title("Smoker vs Non Smoker")

plt.show()
```

```python

x = np.linspace(0, 10, 100)  

fig, ax = plt.subplots(figsize=(5, 3.2))

# sin(x) plot
ax.plot(x, np.sin(x), label='sin(x)', color='#2b6cb0')

# cos(x) plot with dashed line
ax.plot(x, np.cos(x), label='cos(x)', color='#e53e3e', linestyle='--')

ax.set_title('Line Plot')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()
plt.show()
```
