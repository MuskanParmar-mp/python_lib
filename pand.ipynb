```python
import pandas as pd
import matplotlib.pyplot as plt

s = pd.Series([23,45,21,67,34,56],
index=['Jan','Feb','Mar','Apr','May','June'])

s.plot(
    kind='bar',      # chartype
    title='My Chart',    # 
    xlabel='X Label',
    ylabel='Y Label',
    figsize=(10,5),
    color='teal',
    colormap='viridis',
    grid=True,
    legend=True,
    rot=45,
    fontsize=12,

    )
plt.tight_layout()
plt.savefig('chart.png', dpi=150)            # save to file
plt.show()
```
