```python
import numpy as np

# 1 Dimentional array
arr = np.array([1,2,3,4])
print(arr)
print(arr * 2)
print(arr.ndim)


# 2 Dimentional array
arr = np.array([
    [1,2],[3,4]
])
print(arr)
print(arr.ndim)


# 4 Dimentional array
arr = np.array([[[[1, 2, 3, 4]]]])
print(arr)
print(arr.ndim)
```
