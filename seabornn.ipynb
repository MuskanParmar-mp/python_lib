```python
#age_bins = np.arange(0,85,5)
#df['age_group'] = pd.cut(df['age'], bins=age_bins)
#surv_by_age = df.groupby('age_group',observed=True)['survived'].mean()
#midpoints = [interval.mid for interval in surv_by_age.index]
```

```python
df['age_group'] = pd.cut(df['age'],bins = range(0,85,10))

surv_by_age = df.groupby('age_group')['survived'].mean()

plt.Figure()
surv_by_age.plot(kind='line',marker= 'o' , figsize=(5,3)) 
plt.xlabel("age group")
plt.ylabel("survival rate")
plt.title("survival rate of age ")
plt.show()
```
