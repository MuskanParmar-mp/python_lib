### Ques: Find the survival rate by age group

```python
df['age_group'] = pd.cut(df['age'],bins = range(0,85,10))

surv_by_age = df.groupby('age_group')['survived'].mean()

plt.Figure()
surv_by_age.plot(kind='line',marker= 'o' , figsize=(5,3)) 
plt.xlabel("age group")
plt.ylabel("survival rate")
plt.title("survival rate of age ")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,4))
sc = ax.scatter(df['age'], df['fare'], c = df['survived'])
ax.set_title('scatter Plot ---- Age vs Fare ')
ax.set_xlabel('Age')
ax.set_ylabel('Fare (dollar)')
cbar = fig.colorbar(sc, ax = ax, ticks = [0, 1])
cbar.ax.set_yticklabels(['Died', 'Survived'])
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sc = plt.scatter(df["age"],df["fare"], c=df["survived"],cmap="coolwarm")
plt.title("Scatter Plot - Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare (Dollarrrr)")
cbar = plt.colorbar(sc)
cbar.set_ticks([0, 1])
cbar.set_ticklabels(["Died", "Survived"])
plt.show()
```

```python

class_count = df['pclass'].value_counts().sort_index()

plt.figure(figsize=(6, 4))
plt.bar(class_count.index.astype(str), class_count.values)

plt.title("Passenger Count by Class")
plt.xlabel("Passenger Class")
plt.ylabel("Count")

plt.show()
```

```python
grp = df.groupby(['class', 'sex'])['survived'].mean().unstack()

x = np.arange(len(grp.index))
w = 0.35

plt.Figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.bar(x - w/2, grp['male'], w, label = "Male")
plt.bar(x + w/2, grp['female'], w, label = "Female")
plt.xticks(x, grp.index)
plt.title("Survival Rate")
plt.legend()


stack = df.groupby(['class', 'survived']).size().unstack()

plt.subplot(1,2,2)
plt.bar(stack.index , stack[0], label = "died")
plt.bar(stack.index , stack[1], bottom=stack[0], label = "Survived")

plt.title("Outcome counts")
plt.legend()


plt.tight_layout()
plt.show()
```

```python
fig,ax = plt.subplots(2,2, figsize=(8,8))



ax[0, 0].hist(df["age"], bins=20, color="skyblue")
ax[0, 0].set_title("Age Distribution")


ax[0, 1].hist(df["fare"], bins=20, color="orange")
ax[0, 1].set_title("Fare Distribution")

class_counts = df["class"].value_counts()

ax[1, 0].bar(
    class_counts.index, class_counts.values,
    color="green"
)
ax[1, 0].set_title("Passenger by Class")

survival = df.groupby("sex")["survived"].mean()

ax[1, 1].bar(survival.index,survival.values,color="pink"
)
ax[1,1].set_title("survival rate  by gender")

plt.suptitle("titanic dashboard")
plt.tight_layout()
plt.show()

```
