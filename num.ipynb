```python
import numpy as np
```

```python
# 0 Dimention array
arr = np.array(1)
print(arr)
print("0 Dimention = ",arr.ndim)


# 1 Dimention array
arr = np.array(
    [1,2,3,4]
)
print(arr)
print("1 Dimention = ",arr.ndim)


# 2 dimention
arr = np.array([
    [1,2],[3,4]
])
print(arr)
print("2 Dimention = ",arr.ndim)



# 3 dimention
arr = np.array(
    [[[1,2,3]]]
    )
print(arr)
print("3 Dimention = ",arr.ndim)

```

```python
# mean of an array
arr = np.array([10,20,30])
print(np.mean(arr))
```

```python
# multiplication without for loop
arr = np.array([10,20,30])
print(arr * 2)
arr[1]
print("index 1 :", arr[1])
print("minimum :" , np.min(arr)) 
print("maximum :" , np.max(arr))
print("sum :" , np.sum(arr))


# multiplication with for loop
numbers = [10,20,30]
result = []
for i in numbers:
    result.append(i * 2)
print(result)  





```

```python
#  reshape an array and matrix multiplication 
data = np.arange(1,10).reshape(3,3)
data1 = np.arange(1,10).reshape(3,3)
print(data @ data1)
```
